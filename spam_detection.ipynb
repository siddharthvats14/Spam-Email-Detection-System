```python
import numpy as np
import pandas as pd

#We have to split our data set into training data 
#and testing data so using skit learn library function

from sklearn.model_selection import train_test_split

#to convert text data into numberical values for that we're using this

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data collection and pre-processing 
we'll load the data from CSV file to pandas data frame

```python
raw_mail_data = pd.read_csv(r"C:\Users\Revolt's\Desktop\ML Project\Source Code\new_file.csv")
```

```python
print(raw_mail_data)
```

```python
#Replace the NULL values (missing values) with NULL strings
mail_data = raw_mail_data.where((pd.notnull(raw_mail_data)),'')
```

```python
#printing the first 5 rows of dataframe 
mail_data.head()
```

```python
mail_data.shape
```

```python
#Label Encoding
# Spam mail = 0 and ham mail is = 1

mail_data.loc[mail_data['label'] == 'spam', 'label',]=0
mail_data.loc[mail_data['label'] == 'ham', 'label',]=1
```

```python
mail_data.head()
```

```python
#seperating the data as text and label

X = mail_data['text']
Y = mail_data['label']  
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into train data and test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=3)  
```

```python
print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

```python
#Feature Extraction
# basically transform the text data to feature vector that will be the input for the Logistic regression

feature_extraction = TfidfVectorizer(min_df=1, stop_words='english', lowercase=True)

X_train_features = feature_extraction.fit_transform(X_train)
X_test_features = feature_extraction.transform(X_test)

# Convert Y_train and Y_test values to integers
Y_train = Y_train.astype(int)
Y_test = Y_test.astype(int)
```

```python
print(X_train_features)
```

```python
model = LogisticRegression()
```

```python
#Training the Model with the training data
model.fit(X_train_features, Y_train)
```

Evaluating the model

```python
#Perdiction on Trained model

prediction_on_training_data = model.predict(X_train_features)
accuracy_on_training_data = accuracy_score(Y_train, prediction_on_training_data)
```

```python
print('Accuracy:',accuracy_on_training_data)
```

```python
prediction_on_test_data = model.predict(X_test_features)
accuracy_on_test_data = accuracy_score(Y_test, prediction_on_test_data)
```

```python
print('Accuracy:',accuracy_on_test_data)
```
